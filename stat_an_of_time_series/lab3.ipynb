# Лабораторная работа 3

На сетке узлов $\overline{\omega}_h$ найти численное решение смешанной задачи для одномерного уравнения колебаний струны с использованием:
- явной разностной схемы порядка аппроксимации $O(\tau^2 + h^2) $ с $\tau = h = 0.1$;  
- неявной разностной схемы $(\sigma = 0.5)$ порядка аппроксимации $O(\tau^2 + h^2)$ с $\tau = h = 0.1$.

Построить соответствующие разностные схемы. Вычислить погрешность численного
решения путем сравнения результатов с приведенным в задании аналитическим решением $u(x, t)$, т.е. найти $\max_{i,j}|y_i^j - u(x_i,t_j)|$.

Смешанная задача:   
$\begin{cases}
\frac{\partial^2u}{\partial t^2} = \frac{\partial^2u}{\partial x^2} + 5e^{-t}sin(2x) & 0 < x < 1, 0 < t \leq 0.5 \\
u(x, 0) = sin(2x) & 0 \leq x \leq 1 \\
\frac{\partial u(x,0)}{\partial t} = -sin(2x) & 0 \leq x \leq 1 \\
\frac{\partial u(0,t)}{\partial x} = 2e^{-t} & 0 \leq t \leq 0.5 \\
u(1,t) = e^{-t}sin(2) & 0 \leq t \leq 0.5 \\
\end{cases}$   
Точная задача:   
$u(x,t) = e^{-t}sin(2x)$   

Построим явную разностную схему порядка аппроксимации $O(\tau^2 + h^2)$:   
$\begin{cases}
y_{\overline{t}t} = y_{\overline{x}x} + \varphi & x \in \omega_h, t \in \omega_{\tau} \\
y(x, 0) = sin(2x) & x \in \overline{\omega_h} \\
y_t(0) = \tilde{u}_1(x) & x \in \overline{\omega}_h \\
\hat{y}_x(0) = \tilde{\mu}_0(t) & t \in \overline{\omega}_{\tau} \\
y(1,t) = e^{-t}sin(2) & t \in \overline{\omega}_{\tau} \\
\end{cases}$  

$\tilde{u}_1(x) = u_1(x) - \frac{\tau}{2}(u_0''(x) + f(x, 0))$   
$\tilde{\mu}_0(t) = \mu_0(t) - \frac{h}{2}(\check{u}_{\overline{t}t}(0,t) - f(0,t))$

И в индексной форме:   
$\begin{cases}
\frac{y_i^{j+1} - 2y_i^j + y_i^{j-1}}{\tau^2} = \frac{y_{i+1}^{j} - 2y_i^{j} + y_{i-1}^{j}}{h^2} + \varphi_i^{j} & i = \overline{1, N_1 - 1}, j = \overline{1, N_2 - 1} \\
y_i^0 = sin(2x_i) & i = \overline{0, N_1} \\
\frac{y_i^1 - y_i^0}{\tau} = u_1(x_i) - \frac{\tau}{2}(-4sin(2x_i) + f_i^0) & i = \overline{0, N_1} \\
\frac{y_1^{j+1} - y_0^{j+1}}{h} = \mu_0(t_j) - \frac{h}{2}\left(\frac{y_0^{j+1} - 2y_0^j + y_0^{j-1}}{\tau^2} - f_0^j\right) & j = \overline{0, N_2-1} \\
y_{N_1}^{j} = e^{-t_{j}}sin(2) & j = \overline{0, N_2} \\
\end{cases}$  

$\varphi = f$

Алгоритм:   
1) Заполняем нулевой слой:$y_i^0 = u_0(x_i), i = \overline{0, N_1}$   
2) Заполняем первый: 

```python
import numpy as np
import math
from matplotlib import pyplot as plt
```

```python
def u(x, t):
    return math.exp(-t) * math.sin(2 * x)

def f(x, t):
    return 5 * math.exp(-t) * math.sin(2 * x)

def u0(x):
    return math.sin(2 * x)

def u1(x):
    return -math.sin(2 * x)

def u_1(x, tau):
    return u1(x) - tau/2 * (-4 * u0(x) + f(x, 0))

def mu0(t):
    return 2 * math.exp(-t)

def mu_0(t, h, x_arr, t_arr, y, i, j):
    return mu0(t) - h/2 * ((y[j + 1][0] - 2 * y[j][0] + y[j - 1][0])/(tau ** 2) - f(0, t))

def mu1(t):
    return math.exp(-t) * math.sin(2)
```

```python
h = 0.1
tau = 0.1
N1 = int(1 // h)
N2 = int(0.5 // tau)

x = []
for i in range(N1 + 1):
    x.append(h * i)

t = []
for j in range(N2 + 1):
    t.append(tau * j)

u_0 = []
for x_ in x:
    u_0.append(u0(x_))
    
# mu_0_arr = []
# for t_ in t:
#     mu_0_arr.append(mu_0(t_))
    
mu_N2 = []
for t_ in t:
    mu_N2.append(mu1(t_))
    
phi = []
for j in range(N2 + 1):
    phi.append([])
    for i in range(N1 + 1):
        phi[j].append(f(x[i], t[j]))
```

```python
y = []
y.append([u0(x_) for x_ in x])
y.append([y[0][i] for i in range(N1 + 1)])
for i in range(N1 + 1):
    y[1][i] += tau * u_1(x[i], tau)
    
for j in range(1, N2):
    y.append([])
    for i in range(N1 + 1):
        if 0 < i < N1:
            y[j + 1].append(2 * y[j][i] - y[j - 1][i] + (tau ** 2)/(h ** 2) * (y[j][i + 1] - 2 * y[j][i] + y[j][i - 1]) + (tau ** 2) * phi[j][i])
            if i == 1: # <---------------------------------HERE
                y[j + 1][0] = (y[j + 1][1] - h * (mu0(t[j + 1]) - h/2 * (( - 2 * y[j][0] + y[j - 1][0])/(tau ** 2) - f(x[0], t[j + 1]))))/(1 - (h**2)/(2 * tau**2))
        elif i == 0:
            y[j + 1].append(0)
        elif i == N1:
            y[j + 1].append(mu_N2[j + 1])
```

```python
uline = []
for xl in x:
    uline.append(u(xl, t[0]))


plt.plot(x, uline, label='точное решение', linewidth=4)
plt.plot(x, y[0], label='приближённое решение')
plt.show()
```

```python
uline = []
for tl in t:
    uline.append(u(x[0], tl))


plt.plot(t, np.transpose(y)[0], linewidth=4)
plt.plot(t, uline)
plt.show()
```

```python
uline = []
for xl in x:
    uline.append(u(xl, t[4]))


plt.plot(x, uline, label='точное решение', linewidth=4)
plt.plot(x, y[4], label='приближённое решение')
plt.show()
```

```python
uline = []
for tl in t:
    uline.append(u(x[4], tl))


plt.plot(t, np.transpose(y)[4], linewidth=4)
plt.plot(t, uline)
plt.show()
```

Невязка:

```python
diff = []
for j in range(N2 + 1):
    diff.append([])
    for i in range(N1 + 1):
        diff[j].append(abs(y[j][i] - u(x[i], t[j])))

max_j = []
for d in diff:
    max_j.append(max(d))
    
print(max(max_j))
```

Смешанная задача:   
$\begin{cases}
\frac{\partial^2u}{\partial t^2} = \frac{\partial^2u}{\partial x^2} + 5e^{-t}sin(2x) & 0 < x < 1, 0 < t \leq 0.5 \\
u(x, 0) = sin(2x) & 0 \leq x \leq 1 \\
\frac{\partial u(x,0)}{\partial t} = -sin(2x) & 0 \leq x \leq 1 \\
\frac{\partial u(0,t)}{\partial x} = 2e^{-t} & 0 \leq t \leq 0.5 \\
u(1,t) = e^{-t}sin(2) & 0 \leq t \leq 0.5 \\
\end{cases}$   
Точная задача:   
$u(x,t) = e^{-t}sin(2x)$   

Построим неявную разностную схему порядка $O(h^2 + \tau^2)$.   

$\begin{cases}
y_{\overline{t}t} = \frac{1}{2}y_{\overline{x}x} + \frac{1}{2}\hat{y}_{\overline{x}x} + \phi(x, t) & x \in \overline{\omega_h} \\
y_t(0) = \tilde{u}_1(x) & x \in \overline{\omega}_h \\
\hat{y}_x(0) = \tilde{\mu}_0(t) & t \in \overline{\omega}_{\tau} \\
y(1,t) = e^{-t}sin(2) & t \in \overline{\omega}_{\tau} \\
\end{cases}$

$\tilde{u}_1(x) = u_1(x) - \frac{\tau}{2}(u_0''(x) + f(x, 0))$   
$\tilde{\mu}_0(t) = \mu_0(t) - \frac{h}{2}(\check{u}_{\overline{t}t}(0,t) - f(0,t))$

И в индексной форме:   
$\begin{cases}
\frac{1}{2h^2}y_{i - 1}^{j + 1} + \left(-\frac{1}{h^2} - \frac{1}{\tau^2} \right)y_i^{j + 1} + \frac{1}{2h^2}y_{i-1}^{j+1} = \frac{y_i^{j-1} - 2y_i^j}{\tau^2} - \frac{1}{2} \frac{y_{i+1}^{j-1} - 2y_i^{j-1} + y_{i-1}^{j-1}}{h^2} - f_i^j & i = \overline{1, N_1 - 1}, j = \overline{1, N_2 - 1} \\
y_i^0 = sin(2x_i) & i = \overline{0, N_1} \\
y_i^1 - y_i^0 = \tau u_1(x_i) + \frac{\tau^2}{2}(-4sin(2x_i) + f_i^0) & i = \overline{0, N_1} \\
\left(\frac{1}{h} + \frac{h}{2\tau^2}\right)y_0^{j+1} - \frac{1}{h}y_1^{j+1} = \mu_0(t_{j+1}) - \frac{h}{2}\left(\frac{y_0^{j-1} - 2y_0^j}{\tau^2} - f_0^j\right) & j = \overline{0, N_2-1} \\
y_{N_1}^{j} = e^{-t_{j}}sin(2) & j = \overline{0, N_2} \\
\end{cases}$  

```python
A = []
A = np.zeros((N1 + 1, N1 + 1))
for i in range(N1 + 1):
#     A.append([])
    for j in range(N1 + 1):
        if i == 0 and j == 0:
            A[i][j] = (1/h + h/(2 * tau ** 2))
        elif i == 0 and j == 1:
            A[i][j] = -1/h
        elif i == 0 and j != 0 and j != 1:
            A[i][j] = 0
        elif i == N1:
            if j == i:
                A[i][j] = 1
            else:
                A[i][j] = 0
        elif 0 < i < N1:
            if j == i:
                A[i][j] = (-1/h**2 - 1/tau**2)
            elif j == i - 1 or j == i + 1:
                A[i][j] = (1/(2*h**2))

                
y0 = []
for x_ in x:
    y0.append(u0(x_))
    
y1 = []
for i in range(N1 + 1):
    y1.append(y0[i] + tau * u1(x[i]) + tau**2 / 2 * (-4 * math.sin(2 * x[i]) + f(x[i], t[0])))

y0 = np.array(y0)
y1 = np.array(y1)
        
y = []
y.append(y0)
y.append(y1)

# for i in range(N1 + 1):
#     if i == 0:
#         yj.append(yj0)
#     else:
#         yj.append([])
        
for j in range(1, N2):
    b = []
    for i in range(N1 + 1):
        if i == 0:
            b.append(mu0(t[j + 1]) - h/2 * ((y[j-1][0] - 2*y[j][0])/tau**2 - f(x[0], t[j])))
        elif i == N1:
            b.append(mu1(t[j]))
        else:
            b.append((y[j-1][i] - 2*y[j][i])/tau**2 - 1/2 * (y[j-1][i+1] - 2*y[j-1][i] + y[j-1][i-1])/h**2 - f(x[i], t[j - 1]))
            
            
    y.append(np.dot(np.linalg.inv(A), b))
```

```python
uline = []
for xl in x:
    uline.append(u(xl, t[0]))


plt.plot(x, uline, label='точное решение', linewidth=4)
plt.plot(x, y[0], label='приближённое решение')
plt.show()
```

```python
uline = []
for tl in t:
    uline.append(u(x[0], tl))


plt.plot(t, np.transpose(y)[0], linewidth=4)
plt.plot(t, uline)
plt.show()
```
